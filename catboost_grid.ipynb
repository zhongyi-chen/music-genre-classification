```python
from catboost import CatBoost
import numpy as np
import pandas as pd
import random
from catboost import *
from features_preprocessing import partial_dataset, full_dataset 
```

```python
train_data, test_dataset, train_labels = full_dataset()
cat_features = list(range(0, train_data.shape[1]))
print(cat_features)
```

```python
model = CatBoostClassifier(    
    task_type='GPU',
    random_seed=66,
    loss_function='MultiClass',
    eval_metric = 'Accuracy',
    early_stopping_rounds=500,
    )

grid = {
    'learning_rate': [0.055,0.075,0.1],
    'depth':[4,6,8,],
    'bagging_temperature' :[0.4,0.5,0.6,0.8],
    'l2_leaf_reg': [3,6],
    'border_count':[128,255],
    'random_strength':[0.3,0.5,0.7],
    'ctr_target_border_count':[128,255],
    
    
    
}

grid_search_result = model.grid_search(grid, 
                                       X=train_data, 
                                       y=train_labels, 
                                       plot=True)
```
